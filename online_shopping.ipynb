```python
# online shopping cart project

#product catalog (dictionary with project id as key )

product = {
    1: {"name": "Laptop", "price": 50000, "stock": 5},
    2: {"name": "Phone", "price": 20000, "stock": 10},
    3: {"name": "Head_ph", "price": 1500, "stock": 15},
    4: {"name": "Charger", "price": 700, "stock": 20}
}

#Empty cart

cart = {} 
#Fuction
def show_products():
    print("\nAvailable products:")
    print("ID\tName\tPrice\tStock")
    for pid, details in product.items():
        print(f"{pid}\t{details['name']}\t{details['price']}\t{details['stock']}")

#function

def add_to_cart():
    show_products()
    try:
        pid = int(input("Enter product ID to add: "))
        if pid not in products:
            print("Invalid product ID!: ")
            return
        qty = int(input("Enter quantity: "))
        if qty > products[pid]["stock"]:
            print("sorry,not enough stock!")
            return 
            if pid in cart:
                cart[pid] += qty
            else:
                cart[pid] = qty
                product[pid] -= qty
                print(f"{products[pid]['name']} added to cart.")
    except valueError:
        print("Invaild input!")

#function
def remove_from_cART():
    if not cart:
        print("cart is empty!")
        return
    print("\nItems in cart:")
    for pid, qty in cart. items():
        print(f"{pid} - {products[pid]['name']} : {qty}")
    try:
        pid = int(input("Enter product ID to remove: "))
        if pid not in cart:
            print("product not in cart!")
            return
        if qty >= cart[pid]:
            product[pid]["stock"] += qty
            del cart [pid]
        else:
            cart[pid] -= qty
            product[pid]["stock"] += qty
        print("Item updated in cart!")
    except valueError: 
        print("Invaild input!")

#funtion
def view_cart():
    if not cart:
        print("cart is empty!")
        return
    total = 0
    print("\nyour cart:")
    print("Name\tQuantity\tPrice")
    for pid, qty in cart.items():
        name=product[pid]["name"]
        price =product[pid]["price"] * qty
        total += price
        print(f"{name}\t{qty}\t\t{price}")
    print(f"Total Amount: {total}")

#function


def checkout():
    if not cart:
        print("cart is empty!")
        return
    total = 0
    print("\n--- Bill ---")
    with open("receipt.txt","w") as f:
        f.write("----- Receipt -----\n")
        f.write("Name\tQuantity\tPrice\n")
        for pid, qty in cart.items():
            name = product[pid]["name"] * qty
            total += price
            f.write(f"{name}\t{qty}\t{price}\n")
        f.write(f"Total amount to pay: {total}")
        f.write("--------------------------\n")
        print(f"Total amount to pay: {total}")
        print("Receipt saved as 'receipt.txt'")
        cart.clear()#empty the cart ofter checkout


#main menu loop


while true:
    print("\n--- Online shopping cart ---")
    print("1. show products")
    print("2. add to cart")
    print("3. Remove from cart")
    print("4. View cart")
    print("5. checkout")
    print("6. Exit")

    try:
        choice = int(input("Enter your choice"))
        if choice == 1:
            show_products()
        elif choice == 2:
            add_to_cart()
        elif choice == 3:
            remove_from_cart()
        elif choice == 4:
            view_cart()
        elif choice == 5:
            checkout()
        elif choice ==6:
            print("Thank you for shopping! GoodBye!")
            break
        else:
            print("Invalid choice!")
    except valueError:
        print("Please enter a vild number!")
```
